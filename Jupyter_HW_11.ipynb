```python
from sympy import*
from sympy.plotting import plot
init_printing()
x = Symbol("x")
```

```python
f=x**2-6*x+8
grafic = plot(f)
```

```python
solve(f)
```

```python
derivate = diff(f)
```

```python
solve(derivate)
```

```python
roots=solve(derivate)
if len(roots) == 1:
    root = roots.pop()
    if is_increasing(f,Interval(-oo,root)):
        print(f'функция возрастает на промежутке {Interval(-oo, root)}')
    else:
        print(f'функция убывает на промежутке {Interval(-oo, root)}') 
```
